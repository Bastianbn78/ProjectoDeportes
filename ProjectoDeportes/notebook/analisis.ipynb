# Analisis de Datos Deportivos
Este notebook contiene el proceso completo de analisis usando las funciones de limpieza y analisis del proyecto.

```python
import pandas as pd
import sys
import os
from pathlib import Path


sys.path.append(os.path.abspath('..'))

from outputs.src.limpieza import limpiar_deportistas
from outputs.src.analisis import Resumen_datos, Promedio_por_deporte, Deportistas_destacados
```

```python

print('=' * 50)
print('CARGANDO DATOS')
print('=' * 50)

base_path = Path('..').resolve()
csv_path = base_path / 'data' / 'deportistas.csv'
df = pd.read_csv(csv_path)

print(f'Archivo: {csv_path}')
print(f'Registros cargados: {len(df)}')
print(f'Columnas: {list(df.columns)}')
```

```python
# 1. Exploracion inicial del dataset
print('\n' + '=' * 50)
print('EXPLORACION INICIAL')
print('=' * 50)

print('\nPrimeras 10 filas:')
display(df.head(10))

print('\nTipos de datos:')
display(df.dtypes.to_frame('dtype'))

print('\nValores nulos por columna:')
display(df.isnull().sum().to_frame('nulos'))

print(f'\nNumero de filas duplicadas: {df.duplicated().sum()}')

print('\nProblemas detectados en el dataset:')
print('1. Filas mal parseadas (valores concatenados en nombre y NaN en otras columnas).')
print('2. Inconsistencias en etiquetas de deporte (mayusculas, tildes y variantes).')
print('3. Valores faltantes en multiples columnas numericas y categoricas.')
print('4. Posibles outliers en peso_kg y horas_entrenamiento_semana.')
print('5. Filas duplicadas.')
```

```python
# 2. Limpieza de datos
print('\n' + '=' * 50)
print('LIMPIEZA DE DATOS')
print('=' * 50)

df_limpio = limpiar_deportistas(df)
print(f'Registros despues de limpieza: {len(df_limpio)}')

print('\nEstadisticas despues de limpieza:')
display(df_limpio.describe(include='all'))
```

Datos  Limpios:

```python
df_limpio.head(5)
```

```python
# 3. Analisis de datos
print('\n' + '=' * 50)
print('ANALISIS DE DATOS')
print('=' * 50)

resumen = Resumen_datos(df_limpio)
print(f"Total de deportistas: {resumen['total_deportistas']}")
print(f"Promedio de rendimiento: {resumen['promedio_rendimiento']:.2f}")
print(f"Edad promedio: {resumen['edad_promedio']:.2f}")

print('\nPromedio de rendimiento por deporte:')
analisis_por_deporte = Promedio_por_deporte(df_limpio)
display(analisis_por_deporte.sort_values('promedio_rendimiento', ascending=False))

destacados = Deportistas_destacados(df_limpio, umbral=7.0)
print(f"\nNumero de deportistas destacados (rendimiento > 7.0): {len(destacados)}")
if len(destacados) > 0:
    print('\nTop 5 deportistas destacados:')
    display(destacados[['nombre', 'deporte', 'rendimiento_score']].head(5))
```

```python
# 4. Exportar resultados
print('\n' + '=' * 50)
print('EXPORTANDO RESULTADOS')
print('=' * 50)

os.makedirs(base_path / 'outputs', exist_ok=True)


df_limpio.to_csv(base_path / 'outputs' / 'deportistas_limpios.csv', index=False)
print('Datos limpios guardados en outputs/deportistas_limpios.csv')

analisis_por_deporte.to_csv(base_path / 'outputs' / 'analisis_por_deporte.csv', index=False)
print('Analisis por deporte guardado en outputs/analisis_por_deporte.csv')
```
